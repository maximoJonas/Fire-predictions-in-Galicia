```python
import folium
import pandas as pd
import numpy as np
from folium import plugins
```

```python
def inc_map():
    
    año=int(input("Dame  un año entre 2001 y 2014: "))

    datatest = pd.read_csv('fires-all 2001-2015.csv')
    
    datatest["fecha"]=pd.to_datetime(datatest['fecha'])
    datatest["Year"]=pd.DatetimeIndex(datatest["fecha"]).year

    datatest["gastos"].fillna(0,inplace=True)

    datatest=datatest[datatest["gastos"]!=0]

    datatest=datatest[datatest["superficie"]>3]

    datatest["causa_supuesta"].fillna(0,inplace=True)
    datatest=datatest[datatest["causa_supuesta"]!=0]
    datatest=datatest[["lat","lng","gastos","superficie","idprovincia","Year"]]

    datatest.dropna(inplace=True)
    df=datatest

    df=df[(df["idprovincia"]==36) | (df["idprovincia"]==15) | (df["idprovincia"]==32) | (df["idprovincia"]==27)]
    df=df[df["Year"]==año]
    

    df=df[["lat","lng","superficie","gastos"]]
    df.lat=df.lat.astype(float)
    df.lng=df.lng.astype(float)
    heat_df=df[["lat","lng"]]
    
    heat_data=list(zip(df.lat, df.lng))

    m = folium.Map(location=( 42.878212,-8.544844), zoom_start=7,tiles="Stamen Toner", width="%100",height="%100")
    minimap = plugins.MiniMap()
    m.add_child(minimap)

    for lat, lng, superficie,gastos in zip(df["lat"],df["lng"],df["superficie"],df["gastos"]):
        superficie=np.sqrt(superficie)
        
        icone1 = folium.Icon(icon="fire", icon_color="#e67e22", color="red", prefix="fa")
        folium.Marker(location=[lat,lng], popup = f" {superficie} hectareas quemadas. \nUn gasto de:  {gastos}€",icon=icone1).add_to(m)
        
        folium.CircleMarker(location=(lat,lng),radius=superficie*2+10, fill_color='red',).add_to(m)
        
       

    folium.plugins.HeatMap(heat_data).add_to(m)
    
    return m
```

```python
inc_map()
```
